```python
import pandas as pd
import numpy as np
```

```python
genres = pd.read_csv('../../data/raw/genres.csv')
tracks_features = pd.read_csv('../../data/raw/tracks_features.csv')
tracks = pd.read_csv('../../data/raw/raw_tracks.csv')
```

```python
max_bit_rate = int(tracks['track_bit_rate'].max())
max_bit_rate
```

```python
#calculate track quality coefficient
tracks['bit_rate_factor'] = np.round(tracks['track_bit_rate']/max_bit_rate, decimals=2)
tracks['bit_rate_factor']
```

```python
max_interest_rate = int(tracks['track_interest'].max())
max_interest_rate
```

```python
#calculate interest coefficient
# Top Chart (Interest factor > 0.01) x =1
# Mid Chart (Interest factor between 0.01 and 0.001) x=0.6
# Bottom Chart (anything else) x=0.3

tracks['interest_factor'] = tracks['track_interest'] / max_interest_rate
tracks['interest_factor'] = tracks['interest_factor'].apply(lambda x: 1 if x >= 0.01 else 0.6 if 0.001<= x < 0.01 else 0.3)
tracks['interest_factor']
```

```python
#calculate average of interest rank by artist
average_interest_by_artist = np.round(tracks['track_interest'].groupby(tracks['artist_name']).mean())
average_interest_by_artist
```

```python
#calculate average of interest rank by genre
np.round(tracks['track_interest'].groupby(tracks['track_genres']).mean())
```

```python
np.round(tracks['track_interest'].groupby(tracks['track_duration']).mean())
```

```python
s = tracks['track_duration'].str.zfill(5)
tracks['track_time'] = s.str[:2] + s.str[2:] + ':00'
tracks['track_time']
```

```python
#track_length = ['less than 5 minutes', 'between 5 and 7 minutes', 'above 7 minutes']

tracks['track_length'] = tracks['track_time'].apply(lambda x: 'less than 5 minutes' if x < '05:00:00' else 'between 5 and 7 minutes' if '05:00:00'<= x < '07:00:00' else 'above 7 minutes')
tracks.head()
```
